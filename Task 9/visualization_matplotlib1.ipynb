```python
#  Multiple Lines with Different Styles
#  This code creates a plot with three lines: a red dashed line, blue squares, and green triangles, using numpy to generate the x-values.
import numpy as np
t=np.arange(0.,5.,0.2)
plt.plot(t,t,"r--",t,t**2,"bs",t,t**3,'g^')
plt.show()
```

```python

# Plotting with Keyword Strings
# This code creates a scatter plot using data from a dictionary. 
# It uses strings to access data for x ('a'), y ('b'), color ('c'), and size ('d') of the points.

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
#  Plotting with Categorical Variables
# This code creates three subplots (bar, scatter, and line plots) using categorical variables for x-values. The plots share the same x and y data.
names=["group_a","group_b","group_c"]
values=[1,10,100]

plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# Working with Multiple Figures and Axes
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')
plt.show()
```

```python
# Working with Text
# This code creates a histogram with custom labels, title, and a text annotation using LaTeX formatting.
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
# This code creates a plot with an annotation marking the local maximum, using an arrow to point to the specific data point.
ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.ylim(-2, 2)
plt.show()
```

```python
# Adjusting Spacing
# Adjusting Subplot Spacing
# You can adjust the spacing around subplots using subplots_adjust:
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
 for j in range(2):
    axes[i, j].hist(np.random.standard_normal(500), bins=50,
    color="black", alpha=0.5)
fig.subplots_adjust(wspace=0, hspace=0)
```
