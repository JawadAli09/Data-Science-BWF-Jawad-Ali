```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
start_date = "Jan 1, 2020"
end_date = "Mar 31, 2020"
idx = pd.date_range(start_date,end_date)
print(f"Length of index is {len(idx)}")
items_sold = pd.Series(np.random.randint(25,50,size=len(idx)),index=idx)
items_sold.plot()
```

```python
# Adding Two Series Together
additional_items = pd.Series(10, index=pd.date_range(start_date, end_date, freq="W"))
total_items = items_sold.add(additional_items, fill_value=0)
total_items.plot()
```

```python
# Resampling Time Series Data
monthly = total_items.resample("1M").mean()
ax = monthly.plot(kind='bar')
```

```python
# Creating a DataFrame from Series
a = pd.Series(range(1, 10))
b = pd.Series(["I", "like", "to", "use", "Python", "and", "Pandas", "very", "much"], index=range(0, 9))
df = pd.DataFrame({ 'A' : a, 'B' : b })
df
```

```python
# Selecting and Filtering DataFrame Columns and Rows
df['DivA'] = df['A'] - df['A'].mean()
df_filtered = df[df['A'] > 5]
df_filtered
```

```python
# Applying Functions to DataFrame Columns
df['LenB'] = df['B'].apply(len)
df
```

```python
#handling missing values
import numpy as np
example1 = pd.Series([0, np.nan, '', None])
print(example1)
print(example1.isnull())  
```

```python
# Droping null values
example1 = example1.dropna()
print(example1)
```

```python
# Fill null values
example3 = pd.Series([1, np.nan, 2, None, 3], index=list('abcde'))
print(example3.fillna(0))

print(example3.fillna(method='ffill'))

print(example3.fillna(method='bfill'))
```

```python
# Removing Duplicate Data
# Identify and remove duplicate values:
example4 = pd.DataFrame({'letters': ['A','B'] * 2 + ['B'], 'numbers': [1, 2, 1, 3, 3]})
print(example4)
print("finding duplicate values")
print(example4.duplicated())

# Drop duplicates
print("removing duplicate values")
print(example4.drop_duplicates())
print("remove dup letters")
print(example4.drop_duplicates(['letters']))
```

Data cleaning and preparation are critical steps in data analysis. Using Python and Pandas, you can efficiently handle missing values, duplicates, and formatting issues to ensure your data is ready for analysis and modeling.
